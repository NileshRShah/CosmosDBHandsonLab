# Apache Spark to Azure Cosmos DB

## Python Connector
pydocmentdb python library

Pros: 
- It's Python

Cons: 
- Limited data transfer bandwidth

```
# import necessary libraries
import pydocumentdb
from pydocumentdb import document_client
from pydocumentdb import documents
import datetime

# Configuring the connection policy (allowing for endpoint discovery)
connectionPolicy = documents.ConnectionPolicy()
connectionPolicy.EnableEndpointDiscovery 
connectionPolicy.PreferredLocations = ["Central US", "East US 2", "Southeast Asia", "Western Europe","Canada Central"]

# Set keys to connect to Cosmos DB 
host = '<cosmosdb host>'
masterKey = '<cosmosdb key>'
client = document_client.DocumentClient(host, {'masterKey': masterKey}, connectionPolicy)

# Configure Database and Collections
databaseId = '<database>'
collectionId = '<collection>'

# Configurations the Cosmos DB client will use to connect to the database and collection
dbLink = 'dbs/' + databaseId
collLink = dbLink + '/colls/' + collectionId

# Set query parameter
querystr = "SELECT * FROM a"
```

```
# Query documents
query = client.QueryDocuments(collLink, querystr, options=None, partition_key=None)

df = spark.createDataFrame(list(query))
```

```
# manipulate the data
df.show()
```

## Java Connector

Pros: 
- it's Python OR Scala
- more performance than python connector

```
# Configuration Map
tweetsConfig = {
"Endpoint" : "<host>",
"Masterkey" : "<key>",
"Database" : "<database>",
"Collection" : "<collection>", 
"preferredRegions" : "East US",
"SamplingRatio" : "1.0",
"schema_samplesize" : "200000",
"query_custom" : "SELECT * FROM c"
}

# all query options can be found at: https://docs.microsoft.com/en-us/azure/cosmos-db/spark-connector

# Read Tweets
tweets = spark.read.format("com.microsoft.azure.cosmosdb.spark").options(**tweetsConfig).load()
tweets.createOrReplaceTempView("tweets")
```

```
tweets.show()
```

```
# manipulate with pyspark
import pyspark.sql.functions as f
tweets.withColumn('wordCount', f.size(f.split(f.col('text'), ' '))).sort(f.col('wordCount').desc()).show()
```
